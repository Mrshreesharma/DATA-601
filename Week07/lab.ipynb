Using the flights dataframe answer the following questions

1. When do the first and last flights leave each day?

2. When there is a missing value in `dep_time` then this is an indication of canceled flight. Find the number of cancelled flights for each (`carrier`, `origin`) combination.

3. Which carrier has the worst `dep_delay`s? 

4. Which plane (`tailnum`) has the worst on-time record?

5. For each plane, count the number of flights before the first delay of greater than 1 hour.

```python
import pandas as pd 
```

```python
url = 'https://raw.githubusercontent.com/mguner/teaching/main/datasets/nyc_flights/nyc_flights.csv'

## read the file
flights = pd.read_csv(url)
```

```python
flights
```

```python
flights.groupby(['day']).agg({'dep_time':'min'})
```

```python
flights.groupby(['day']).agg({'dep_time':'max'})
```

```python
cancelled = flights[flights['dep_time'].isnull()]
cancelled.fillna(value = 1,inplace = True)
```

```python
cancelled.groupby(['carrier','origin']).agg({'dep_time':'count'})
```

```python
delays = flights.groupby(['carrier']).agg({'dep_delay':'mean'})
```

```python
delays.sort_values(by='dep_delay' ,ascending=False)
```

```python
late = flights.groupby(['tailnum']).agg({'arr_delay':'max'})
late.sort_values(by='arr_delay' ,ascending=False)
```

```python
delay = flights[flights['dep_time']< 60]
delay.groupby('tailnum').agg(delay = ('dep_delay','count')).sort_values(by= 'delay')
```
