```python
import pandas as pd

nba_players = pd.read_csv('https://raw.githubusercontent.com/msaricaumbc/DS601_Fall21/main/Week02/data/all_seasons.csv', index_col = 0)

height = nba_players['player_height']
points = nba_players['pts']
```

```python
import matplotlib.pyplot as plt
plt.hist(height)
plt.title('NBA Players Height')
plt.xlabel('Number of Players')
plt.ylabel('Height')
plt.show()
```

```python
plt.boxplot(points)
plt.title('Points Only')
plt.xlabel('Number of Players')
plt.ylabel('Points')
plt.show()
```

```python
plt.plot(height,points,'o',color='black')
plt.title('Height VS Points')
plt.xlabel('Height')
plt.ylabel('Points')
plt.show()
```
